```python
url="https://www.espncricinfo.com/auction/ipl-2025-auction-1460972"
```

```python
import requests
```

```python
A=requests.get(url)
```

```python
A
```

```python
from bs4 import BeautifulSoup as bs
```

```python
B=bs(A.text,"html.parser")
```

```python
B
```

```python
C=B.find("tbody")
```

```python
C
```

```python
D=C.find_all("tr")
```

```python
D
```

```python
column=D[0].find_all("td")
```

```python
column
```

```python
column[0].text
```

```python
column[1].text
```

```python
player=[]
batbowl=[]
base=[]
price=[]
for r in D[0:]:
    column=r.find_all("td")
    player.append(column[0].text)
    batbowl.append(column[1].text)
    base.append(column[2].text)
    price.append(column[3].text)
```

```python
dict={"Player":player,"BatBowl":batbowl,"Base":base,"Price":price}
```

```python
import pandas as pd 
```

```python
CSK=pd.DataFrame(dict)
```
